Correlation between median k-mer count and read coverage calculated from mapping, from both simulated data and genomic & transcriptomic data.

```python
import itertools, gzipdef load_cmp_file(filename, limit=None):    if filename.endswith('.gz'):        fp = gzip.open(filename)    else:        fp = open(filename)    if limit:        lines = [ line.split() for i, line in itertools.izip(range(limit), fp) ]    else:        lines = [ line.split() for line in fp ]    lines = [ (float(a[0]), float(a[1])) for a in lines ]    print 'loaded %d lines' % len(lines)    return numpy.array(lines)
```

```python
genome_counts = load_cmp_file(datadir + 'genome-reads.fa.counts.cmp', limit=num_points_to_plot)ends_counts = load_cmp_file(datadir + 'ends.cmp')
```
